```python
import ee
ee.Initialize()
from IPython.display import Image
```

```python
AOI = ee.Feature(ee.Geometry.Rectangle(26.313846, 62.413351, 26.403882, 62.488352))
#LT4_L1T = ee.ImageCollection('LANDSAT/LT4_L1T')
LT4_L1T = ee.ImageCollection('MODIS/051/MCD12Q1')
```

```python
image = ee.Image(LT4_L1T.first())
print(image.getInfo())
print(image.bandNames().getInfo())
```

```python
Image(url=image.getThumbURL({'format': 'jpg', 'size': '600'}))
```

```python
def clip_fun(img):
    return img.clip(AOI)

def attr_fun(img):
    #return ee.Date(ee.Image(img).get('system:time_start'))
    return ee.Image(img).get('CLOUD_COVER')

def print_cloud(img):
    return ee.Image(img).getInfo()['properties']['CLOUD_COVER']

# Get images between 1.6. and 31.8.
LT4_L1T_summer = LT4_L1T.filter(ee.Filter.dayOfYear(153, 244))
print LT4_L1T_summer.size().getInfo()

# Clip imageCollection with AOI
LT4_L1T_AOI = LT4_L1T_summer.map(clip_fun)

# Get only images without clouds
LT4_L1T_cloudless = LT4_L1T_AOI.filter(ee.Filter.eq('CLOUD_COVER', 0))
print LT4_L1T_cloudless.size().getInfo()

# Create a list of metadata (?) from the cloudless images that is as long as number of cloudless images
imagelist = LT4_L1T_cloudless.toList(LT4_L1T_cloudless.size().getInfo())
print imagelist.get(0).getInfo()
print imagelist.get(0).getInfo()['properties']['CLOUD_COVER']

# There's a strange problem here. If you print the getInfo of the list with the cloud-filtered imagery everything 
# works OK, if you do it with the cloud cover info selected then you get an internal server error.
```
